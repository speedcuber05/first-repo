```python
import rawpy
import numpy as np

path = r"C:\Users\devan\Downloads\IMG_20251227_154935.dng"

with rawpy.imread(path) as raw:
    raw_img = raw.raw_image.astype(np.float32)

    black = raw.black_level_per_channel[0]
    white = raw.white_level

    brightness = raw_img - black
    brightness = np.clip(brightness, 0, None)

    brightness_norm = brightness / (white - black)
    print(brightness_norm.size)
```
