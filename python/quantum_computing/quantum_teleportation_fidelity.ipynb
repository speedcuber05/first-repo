```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0,1,100)


F = (2*p + 1)/3

plt.plot(p, F, label="Teleportation Fidelity")


plt.axhline(2/3, linestyle="--", label="Classical Limit (2/3)")


plt.axvline(1/3, linestyle=":", label="Entanglement Threshold (1/3)")
plt.axvline(1/2, linestyle=":", label="Quantum Advantage (1/2)")
plt.axvline(1/np.sqrt(2), linestyle=":", label="Bell Violation (~0.707)")

plt.xlabel("Werner parameter p")
plt.ylabel("Average Teleportation Fidelity")
plt.title("Teleportation Fidelity vs Entanglement Quality")

plt.legend()
plt.show()
```
