# importing the modules

```python
import numpy as np
import pandas as pd
```

# setting up the file

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.width',750)


filepath=r"C:\Users\devan\OneDrive\Documents\Learning_code\first-repo\male_players.csv"
data=pd.read_csv(filepath)

all_df=data.drop(columns=['url','Rank','Unnamed: 0.1', 'Unnamed: 0'])

# all_df is the base df we are going to use, I dropped all the useless columns, rank will be added later
# as the index
```

### setting up goalkeeper data



```python
gk_raw_df=all_df[all_df["Position"]=="GK"]
gk_df=gk_raw_df[['Name','OVR','GK Diving' , 'GK Handling' , 'GK Kicking' , 'GK Positioning','GK Reflexes',
                 'Weak foot','Skill moves','Preferred foot','Height','Weight','Age','Nation','League',
                 'Team','play style']].copy()

gk_df.index=np.arange(1,len(gk_df)+1)
gk_df.index.name = "Rank"

print(gk_df.head())
```

### setting up outfield data



```python
outfield_raw_df=all_df[all_df['Position']!="GK"]
outfield_df=outfield_raw_df.drop(columns=['GK Diving' , 'GK Handling' , 'GK Kicking' , 'GK Positioning','GK Reflexes']).copy()

outfield_df.index=np.arange(1,len(outfield_df)+1)
outfield_df.index.name = "Rank"
print(outfield_df)
```

# adding some new columns

```python
# adding a column which adds all the stats of the goalkeeper
Stat_Sum= gk_df["GK Diving"] + gk_df["GK Handling"] + gk_df["GK Kicking"] + gk_df["GK Positioning"] + gk_df["GK Reflexes"]

col_position=gk_df.columns.get_loc("GK Reflexes") + 1
gk_df.insert(col_position,"Stat Sum",Stat_Sum)

print(gk_df)
```

```python
# adding a column which will count the number of playstyles a player has, and set the height to an integer in cms
def count_playStyles(df):
    count_ps=np.empty(len(df))
    index=0
    for ps in df['play style']:
        if pd.notna(ps):
            count_ps[index]=len(ps.split(','))
        else:
            count_ps[index]=0
        index+=1
    return count_ps

def get_height(df):
    hts=np.empty(len(df))
    index=0
    for ht in df["Height"]:
        hts[index]=int(ht[0]+ht[1]+ht[2])
        index+=1
    return hts

gk_df["Height"]=get_height(gk_df)
gk_df["no. of play style"]= count_playStyles(gk_df)
gk_df["Total Stats"]= gk_df['Stat Sum'] + gk_df['Height'] + gk_df['no. of play style']
gk_df['Height'] = gk_df['Height'].astype(int)

gk_df

```

```python
# doing the same for outfield players too
outfield_df["Height"]=get_height(outfield_df)
outfield_df["no. of play style"]= count_playStyles(outfield_df)

print(outfield_df)
```

### normalizing free kicks from 0 - 1

```python
def normalization(column):
    outfield_df[f"{column} normalized"]=(outfield_df[column] - outfield_df[column].min())/(outfield_df[column].max()-outfield_df[column].min())
normalization("Free Kick Accuracy")
```

# checking if the ovr ratings are the averages of their respective stats

```python
mismatch_pac=(outfield_df['PAC'] - outfield_df['Acceleration']/2 - outfield_df['Sprint Speed']/2)
mismatch_pac=mismatch_pac[mismatch_pac.abs()>0.5]
result = pd.DataFrame({
    'Name': outfield_df.loc[mismatch_pac.index, 'Name'],
    'PAC': outfield_df.loc[mismatch_pac.index, 'PAC'],
    'PAC_mismatch': mismatch_pac
})
print(result.shape)

    
```

# sorting data with groupby

```python
# checking the highest avg stamina by the position
position_df=all_df.groupby('Position')
avg_sta_position=position_df["Crossing"].mean().sort_values(ascending=False)
print(avg_sta_position)
```

```python
# checking the best nation by the avg ovr

filter_all_df=all_df[all_df['OVR']>73]
nation_df=filter_all_df.groupby('Nation')
avg_ovr_nation=nation_df['OVR'] \
.agg(['mean','count']) \
.query('count > 11') 
avg_ovr_nation['mean'] = avg_ovr_nation['mean'].round().astype(int)
print(avg_ovr_nation.sort_values(by=['mean','count'],ascending=False))
```
